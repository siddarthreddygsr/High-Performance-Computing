```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('q3.txt', delimiter = " ",header = None)
matdata_2048 = pd.read_csv('q31_2048.txt', delimiter = " ", header = None)
```

```python
data.head()
```

```python
data.columns = ['Time', 'Size', 'Power', 'Thread']
```

```python
matdata_512 = data[data['Size'] == 512]
matdata_1024 = data[data['Size'] == 1024]
# matdata_2048 = data[data['Size'] == 2048]
```

```python
print(matdata_512.shape)
print(matdata_1024.shape)
print(matdata_2048.shape)
```

```python
print(matdata_1024)
```

```python
matdata_2048.columns =  ['Time', 'Size', 'Power', 'Thread']
```

```python
matdata_2048
```

```python
matdata_1024.reset_index(inplace = True)
matdata_2048.reset_index(inplace =True)
```

```python
del matdata_1024['index']
del matdata_2048['index']
```

```python
#For each power --> for each Thread --> 
```

```python
unique_threads = data['Thread'].unique()
unique_threads
```

```python
unique_power = data['Power'].unique()
unique_power
```

```python
   
             
for power in unique_power:
    x = []
    y = []
    for i in range(len(matdata_512)):
        if matdata_512.iloc[i]['Power'] == power:
            x.append(matdata_512.iloc[i]['Time'])
            y.append(matdata_512.iloc[i]['Thread'])
#     plt.title('Time taken for the 512 matrix for calculating power of ' + str(power+1))
    plt.plot(y,x, marker = "o", label = str(power+1) + " power")
    plt.legend()
    
    plt.xlabel('Number of Threads')
    plt.ylabel('Time taken to run the Code')
plt.title("Number of Threads vs Time (in s) [512x512 matrix]")
plt.show()
```

```python
for power in unique_power:
    x = []
    y = []
    for i in range(len(matdata_1024)):
        if matdata_1024.iloc[i]['Power'] == power:
            x.append(matdata_1024.iloc[i]['Time'])
            y.append(matdata_1024.iloc[i]['Thread'])
#     plt.title('Time taken for the 1024 matrix for calculating power of ' + str(power+1))
    plt.plot(y,x,marker = "o", label = str(power+1) + " power")
    plt.xlabel('Number of Threads')
    plt.ylabel('Time taken to run the Code')
    
plt.title("Number of Threads vs Time (in s) [1024x1024 matrix]")
plt.show()
```

```python
for power in unique_power:
    x = []
    y = []
    for i in range(len(matdata_2048)):
        if matdata_2048.iloc[i]['Power'] == power:
            x.append(matdata_2048.iloc[i]['Time'])
            y.append(matdata_2048.iloc[i]['Thread'])
    plt.title('Time taken for the 2048 matrix for calculating power of ' + str(power+1))
    plt.plot(y,x,marker = "o", label = str(power) + " power")
    plt.xlabel('Number of Threads')
    plt.ylabel('Time taken to run the Code')

plt.title("Number of Threads vs Time (in s) [2048x2048 matrix]")
plt.show()
```

```python
for power in unique_power:
    x_512 = []
    y_512 = []
    x_1024 = []
    y_1024 = []
    x_2048 = []
    y_2048 = []
    for i in range(len(matdata_2048)):
        if matdata_2048.iloc[i]['Power'] == power:
            x_512.append(matdata_512.iloc[i]['Time'])
            y_512.append(matdata_512.iloc[i]['Thread'])
            x_1024.append(matdata_1024.iloc[i]['Time'])
            y_1024.append(matdata_1024.iloc[i]['Thread'])
            x_2048.append(matdata_2048.iloc[i]['Time'])
            y_2048.append(matdata_2048.iloc[i]['Thread'])
            
    plt.title('Time taken for different size matrices for calculating power of ' + str(power+1))
    plt.plot(y_1024,x_1024,marker = "o", label = '1024x1024 matrix size')
    plt.plot(y_512,x_512,marker = "o", label = '512x512 matrix size')
    plt.plot(y_2048,x_2048,marker = "o", label = '2048x2048 matrix size')
    plt.legend()
    plt.xlabel('Number of Threads')
    plt.ylabel('Time taken to run the Code')
    plt.show()
```

```python
q3_df = pd.concat([matdata_512, matdata_1024, matdata_2048])
```
